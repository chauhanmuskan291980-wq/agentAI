### Messages
Messages are the fundamental unit of context for models in Langchain. They represent the input and output of models , carrying both the content and metadata needed to represet the state of conversation , when interacting with an LLM. Messages are objects that contain:

- Role : Ideentifies the message type(e.g. system , user)
- Content : Represents rge actualcontent of the message(like text , images , audio , documents etc,)
- Metadata : Optional Fields such as response information , messsage IDs and token usage

Langchain provides a standard message type that works across all model provides , ensuring consistent behavior regardless of the model being called.


```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
model.invoke("Please tell me what is ML")
```

### Text Prompts

Text prompts are strings - ideal for staightforword generation tasks where you don't need to retain conversation history.

```python
model.invoke("What is langchain")
```

### Use text prompts when:
- You have a single , standalone request
- You don't need conversation history
- You want minimal code complexity

### Message Prompts
Alternatively , you can pass in a list of messages to the model by provinding a list of message objects.
<b>- Message Type:</b>
- System Message : Tells the model how to behave and provide context for interactions
- Human Message : Represents user input and interacyion with the model
- AI message : Response generated by the model , including text content m tool calls and metadata
- Tool Message - Represents the output of tool calls

### System Message

A SystemMessage represent an intial set of instructions that primes the model's behavior , You can use a system message to set the tome define the model's role , and establich guidelines for responses.


### Human Message
A HumanMessage represents user input and interactions. They can contain text , image m audio , files and any other amount of multimodel content.

### AI Message
An AI Message represents the putput of a model invocation , They can include multimodel data , tool calls and provider-specfic metadat that you can later access.

### Tool Message
For models that supports tool callimng . AI messages can contain tool calls. Tool message are used to pass the results of a single tool exectution back to the model 



```python
from langchain.messages import SystemMessage , HumanMessage, AIMessage

messages = [
    SystemMessage("You are a poetry expert"),
    HumanMessage("write a poem on artifical intelligence")
]

response = model.invoke(messages)
response.content
```

```python
system_msg = SystemMessage("You are a helpful coding assistant.")

message = [
    system_msg,
    HumanMessage("How do I create a REST API ?")
]
response = model.invoke(message)
print(response.content)
```

```python
## Detailed info to the LLM throught system Message 
from langchain.messages import SystemMessage , AIMessage , HumanMessage

system_msg = SystemMessage("""
You are a senior python developer with expertise in web frameworks.
                           Always provides code exapmle and explain you reasoning.
                           Be concise nut thorough in your explanations.
                           """)

message = [
    system_msg,
    HumanMessage("How do I create a REST API?")
]

response = model.invoke(message)
print(response.content)
```

```python
## Message Metadata

human_msg = HumanMessage(
    content = "hello!",
    name="alice",    ## Optional : Identofy diffrent users
    id = "msg_123",  ## Optional : unique identifier for tracing
)
```

```python
response = model.invoke([
    human_msg,
])
response
```

```python
from langchain.messages import AIMessage , SystemMessage , HumanMessage

ai_msg = AIMessage("I'd be happy to help you with that question!")

messages = [
    SystemMessage("You are a helpful assistent"),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Great! What's 2+2")
]

response = model.invoke(messages)
print(response.content)
```

```python
from langchain.messages import AIMessage
from langchain.messages import ToolMessage

ai_message = AIMessage(
    content = [],
    tool_calls =
    [{
            "name":"get_weather",
            "args":{"location":"San Francisco"},
            "id":"call_123"
        }]
)

## Execute tool and create result message

weather_result = "Sunny 72 F"
tool_message = ToolMessage(
    content = weather_result,
    tool_call_id = "call_123"
)

## Continue conversation 
messages = [
    HumanMessage("What's the weather in San Frencisco?"),
    ai_message,
    tool_message
]
response = model.invoke(message)
print(response.content)
```
