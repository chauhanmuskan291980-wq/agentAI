### Steaming and Batch 
- Streaming 
Most models can steam their output content while it is being generated. By displaying output progressively , steaming significantly improves user esperience , aprticularly for longer response, Calling Steam() return and iterator that yields output chunks as they are peoduced, You can use a loop to process each chunk in real-time.


```python
model.invoke("Write me a 200 words paragraph on AI")
```

```python
for chunk in model.stream("Write me a 200 words paragraph on AI"):
    print(chunk.text, end="", flush=True)
```

### Batch
Batching a collection of independent requests to a model can significantly improve performance and reduce costs , as the processing can be done in parallel:

```python
responses = model.batch([
    "WHy do parrots have colorful feathers?",
    "How do airplane fly?",
    "WHat is quantum computing?"
])
 
for response in responses:
 print(response)
```

```python
responses = model.batch([
    "WHy do parrots have colorful feathers?",
    "How do airplane fly?",
    "WHat is quantum computing?"],
    config={
        "max_concurrency": 5 ,  # Limit to parallel calls
    }
    
    )

responses
```
